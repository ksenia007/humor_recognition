```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
colors = ["amber", "windows blue", "greyish", "faded green", "dusty purple"]
sns.set_palette(sns.xkcd_palette(colors))
import re
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
puns = np.load('results/embeddings/puns_embeddings.npy', allow_pickle=True)
onelines = np.load('results/embeddings/oneliners_embeddings.npy', allow_pickle=True)
humi = np.load('results/embeddings/humicroedit_embeddings.npy', allow_pickle=True)
all_e = np.concatenate([puns, onelines, humi])
```

```python
pca = PCA(n_components=4)
pca_t = pca.fit_transform(all_e)
pca.explained_variance_ratio_
```

```python
plt.scatter(pca_t[:,0], pca_t[:,1], alpha=0.1)
```

```python
X_embedded = TSNE(n_components=2, perplexity=20, n_iter=500).fit_transform(pca_t)
```
